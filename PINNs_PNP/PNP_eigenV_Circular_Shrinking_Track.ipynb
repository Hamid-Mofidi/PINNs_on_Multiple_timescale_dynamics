## Circular Shrinking Track based on the eigen vectors in PNP model:

\begin{equation}
\begin{aligned}
u =& r  \Big( \cos(t) -  \sin(t) \Big), \\
z_1 c_1 + z_2 c_2 =&  r  \Big(\cos(t) + \sin(t) \Big),
\end{aligned}
\end{equation}
where
$$
r = r_0 * e^{-\lambda t}
$$
with $r_0 =1$ and $\lambda = 0.2$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
r_0 = 1        # Initial radius
lambda_ = 0.2  # Rate of shrinking
omega = 1     # Angular velocity

# Define the range of t
t = np.linspace(0, 10 * np.pi, 1000)

# Calculate r and theta as functions of t
r = r_0 * np.exp(-lambda_ * t)


# Convert polar coordinates to Cartesian coordinates
u = r * (np.cos(t) -  np.sin(t))
z1c1z2c2 = r * (np.cos(t)  + np.sin(t) )

# Create the plot
plt.figure(figsize=(8, 8))
plt.plot(u, z1c1z2c2)
plt.xlabel('$u$')
plt.ylabel('$z_1c_1 + z_2c_2$')
plt.title('Circular Shrinking Track')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

plt.gca().invert_xaxis()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters
r_0 = 1        # Initial radius
lambda_ = 0.4 # Rate of shrinking
omega = 1      # Angular velocity

# Define the range of t
t = np.linspace(0, 10 * np.pi, 1000)

# Calculate r as a function of t
r = r_0 * np.exp(-lambda_ * t)

# Convert polar coordinates to Cartesian coordinates
w = np.ones_like(t)
u = r * (np.cos(t) -  np.sin(t))
z =  r * (np.cos(t) + np.sin(t))

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(w, u, z, label='3D Shrinking Track')
ax.set_xlabel('w')
ax.set_ylabel('u')
ax.set_zlabel('$z_1c_1 + z_2c_2$')
ax.zaxis.labelpad = -1
ax.set_title('3D Plot of Circular Shrinking Track with z = 1')
ax.legend()
ax.invert_xaxis()  # Reverse the u-axis
ax.view_init(elev=20)
plt.show()
```
